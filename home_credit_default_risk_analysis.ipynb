#### Differences among Targets

```python
def ttest_apply(x, var_name='TARGET', groups=target_groups, mean_std_nobs=['mean', 'std', 'count']):
    """Apply ttest for numerical values and get p value
    Parameters
    ----------
    x : pd.Series like list, with name
    
    Returns
    -------
    p_val :  p value
    """

    ttest_ans = ttest_ind_from_stats(mean1=x[var_name + '_' + str(groups[0]) + '_' + mean_std_nobs[0]], 
                                     std1=x[var_name + '_' + str(groups[0]) + '_' + mean_std_nobs[1]], 
                                     nobs1=x[var_name + '_' + str(groups[0]) + '_' + mean_std_nobs[2]],
                                     mean2=x[var_name + '_' + str(groups[1]) + '_' + mean_std_nobs[0]],
                                     std2=x[var_name + '_' + str(groups[1]) + '_' + mean_std_nobs[1]], 
                                     nobs2=x[var_name + '_' + str(groups[1]) + '_' + mean_std_nobs[2]],
                                     equal_var = False, alternative = 'two-sided')

    return([ttest_ans.pvalue, ttest_ans.statistic])

agg_df = all_data[num_cols.to_list() + ['TARGET']].groupby('TARGET').agg(['mean', 'std', 'count']).transpose()
agg_df.reset_index(inplace=True)
agg_df = pd.pivot_table(agg_df, index=['level_0'], columns=['level_1'])
agg_df.columns = ['TARGET_'+str(x[0])+'_'+x[1] for x in agg_df.columns]
agg_df.reset_index(inplace=True)

agg_df[['pvalue', 'statistics']] = pd.DataFrame(agg_df.apply(ttest_apply, axis=1, result_type = 'expand'))

agg_df = agg_df.sort_values(['pvalue'], ascending=False)
agg_df[agg_df['pvalue'] < 0.05].tail(10).sort_values(['pvalue'], ascending=True)
```

```python
# cat_col_index = [list(X_df.columns).index(col) for col in cat_cols]
# categories_col = []
# total_cat_count = 0
# for col in cat_cols:
#     col_cat_unique = pd.unique(df[col])
#     if any(pd.isnull(col_cat_unique)):
#         col_cat_unique = np.append(col_cat_unique[~pd.isnull(col_cat_unique)], np.nan)
#     categories_col.append(col_cat_unique)
#     total_cat_count += len(col_cat_unique)

# onehot_encoder = OneHotEncoder(categories = categories_col, drop=None, sparse_output=True, handle_unknown='ignore')
# onehot_encoder.fit(df[cat_cols])

# cat_onehot_df = onehot_encoder.transform(df[cat_cols])
# cat_onehot_df

# X_onehot = df.drop(list(cat_cols), axis=1)
# X_onehot = hstack((X_onehot, cat_onehot_df))

# np.concatenate((X_onehot, cat_onehot_df), axis=1)
# pd.concat([X_onehot, cat_onehot_df], axis=1)

def onehot_encoder_process(data, category_columns):
    """Apply one hot encoder with manual labeled categories
    Parameters
    ----------
    data : pd.DataFrame
    category_colulmns: list, contains names of all categorical columns
    
    Returns
    -------
    ans[0]: onehot_mat, scipy.sparse matrix
    ans[1]: onehot_encoder, sklearn.preprocessing.OneHotEncoder
    ans[2]: onehot_
    """
    # cat_col_index = [list(data.columns).index(col) for col in category_columns]

    # Summarize the categorical data
    categories_col = []
    total_cat_count = 0
    for col in category_columns:
        col_cat_unique = pd.unique(data[col])
        if any(pd.isnull(col_cat_unique)):
            col_cat_unique = np.append(col_cat_unique[~pd.isnull(col_cat_unique)], np.nan)
        categories_col.append(col_cat_unique)
        total_cat_count += len(col_cat_unique)

    # OneHotEncoder
    onehot_encoder = OneHotEncoder(categories = categories_col, drop=None, sparse_output=True, handle_unknown='ignore')
    data_cat = data[category_columns] # pd.DataFrame, keep feature names
    onehot_encoder.fit(data_cat)
    data_cat_onehot = onehot_encoder.transform(data_cat)

    # Value Check, number of columns matches
    print('Compatiblility Check: ', data_cat_onehot.shape[1] == total_cat_count)

    # Combine with non-categorical data
    data_onehot = data.drop(category_columns, axis=1)
    onehot_mat = hstack((data_onehot, data_cat_onehot))

    # column names
    column_names = list(data_onehot.columns) + list(onehot_encoder.get_feature_names_out())
    
    # results
    return([
        onehot_mat,
        onehot_encoder,
        column_names
    ])

# test
X_onehot, onehot_encoder, X_onehot_colnames = onehot_encoder_process(df.drop([id_cols, target_cols], axis=1), cat_cols)
```

### LightGBM & hyperopt

```python
result_ho = pd.DataFrame()
i = 0

# define an objective function
## will return the score to optimise
def objective(space):
    
    global i
    global result_ho
    
    #Creating Lightgbm DataFrame
    lgb_train = lgb.Dataset(X_train_selected, label = y_train)
    lgb_val = lgb.Dataset(X_test_selected, label = y_test)

    #Parameters of model
    params = {
        'num_leaves': int(space['num_leaves']),
        'max_depth': int(space['max_depth']),
        'learning_rate': space['learning_rate'],
        'objective': 'binary',
        'metric': 'auc',
        "boosting": "gbdt",
        'feature_fraction' : space['feature_fraction'],
        'max_bin' : int(space['max_bin']),
        'min_data_in_leaf': int(space['min_data_in_leaf']),
        "min_data_in_bin": int(space['min_data_in_bin']),
        "bagging_freq": 20,
        "random_seed": 2019,
        "lambda_l1": space['lambda_l1'],
        "lambda_l2": space['lambda_l2'],
        'pos_bagging_fraction' : space['pos_bagging_fraction'],
        'neg_bagging_fraction' : space['neg_bagging_fraction'],
        'verbose': -1
    }

    evals_result = {}
    
    clf = lgb.train(params, lgb_train, 20000, valid_sets=lgb_val,
                    valid_names='val',
                    callbacks=[lgb.early_stopping(stopping_rounds=5)])
                    # evals_result=evals_result)
    gc.collect()
    
    result = pd.DataFrame(clf.params, index=[0])
    
    ## Calculating AUC
    pred_train = clf.predict(X_train_selected)
    pred_val = clf.predict(X_test_selected)

    gc.collect()
    train_auc = roc_auc_score(y_train, pred_train)#, num_iteration=clf.best_iteration)
    val_auc = roc_auc_score(y_test, pred_val)#, num_iteration=clf.best_iteration)
    gc.collect()

    score = (abs(train_auc - val_auc) + 1)/((1+val_auc)*(1+val_auc))
    
    result["train_auc"] = train_auc
    result["val_auc"] = val_auc
    result["train_test_diff"] = train_auc - val_auc
    result["n_estimators"] = clf.best_iteration
    result["score"] = score
    
    del clf
    
    result_ho = pd.concat([result_ho,result], ignore_index=True)
    i = i+1
    
    return (score)

# define a search space

## Hyperparameter space
## Space is selection of data point from the given distribution
## Distribution is defined for every hyperparameter seperately
space = {
    'num_leaves': hp.quniform('num_leaves', 2, 24, 1), # Uniform integer between 2 and 24
    'max_depth': hp.quniform('max_depth', 2, 12, 1), # Uniform integer between 2 and 12
    'learning_rate': hp.uniform('learning_rate', 0.005, 0.015), # Values between 0.005 to 0.015
    'feature_fraction' : hp.uniform('feature_fraction', 0.1, 1), # Values between 0.1 to 1
    'max_bin' : hp.quniform('max_bin', 10, 100, 10),
    'min_data_in_leaf' : hp.quniform('min_data_in_leaf', 25, 1000, 25),
    'lambda_l1' : hp.uniform('lambda_l1', 0, 50),
    'lambda_l2' : hp.uniform('lambda_l2', 0, 50),
    'min_data_in_bin' : hp.quniform('min_data_in_bin', 5, 100, 5),
    'pos_bagging_fraction' : hp.uniform('pos_bagging_fraction', 0.1, 1),
    'neg_bagging_fraction' : hp.uniform('neg_bagging_fraction', 0.1, 1)
    }
# minimize the objective over the space
best=fmin(fn = objective, # function to optimize
          space = space, # space from which hyperparameter to be choosen
          algo = tpe.suggest, # optimization algorithm, hyperopt will select its parameters automatically
          max_evals = 50,
          rstate = np.random.default_rng(7)
         )

result_ho.to_csv(result_dir_loc+'/'+'hyperopt_results.csv', index=False)
```
